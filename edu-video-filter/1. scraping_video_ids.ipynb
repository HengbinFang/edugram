```python
ids = {}
```

```python
import re

def subtitles(lines):
    lines = lines.split('\n')
    text = ''
    for line in lines:
        if re.search('^[0-9]+$', line) is None and re.search('^[0-9]{2}:[0-9]{2}:[0-9]{2}', line) is None and re.search('^$', line) is None:
            text += ' ' + line.rstrip('\n')
        text = text.lstrip()
    return text

    
```

```python
import requests

# The way I fetched the videos were very hacky. Needs cookies. If you want to try this, grab your cookies from Instagram.com
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:120.0) Gecko/20100101 Firefox/120.0',
    'Accept': '*/*',
    'Accept-Language': 'en-CA,en-US;q=0.7,en;q=0.3',
    # 'Accept-Encoding': 'gzip, deflate, br',
    'X-CSRFToken': '',
    'X-IG-App-ID': '',
    'X-ASBD-ID': '',
    'X-IG-WWW-Claim': '',
    'X-Requested-With': 'XMLHttpRequest',
    'Alt-Used': 'www.instagram.com',
    'Connection': 'keep-alive',
    'Referer': 'https://www.instagram.com/explore/tags/togetherforgood/',
    'Cookie': "", 
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-origin',

}
session = requests.Session()
session.headers.update(headers)

def get_tags(tag):
    params = {
        'tag_name': tag,
    }

    response = session.get('https://www.instagram.com/api/v1/tags/web_info/', params=params)
    try:
        data=response.json()

        for x in data["data"]["top"]["sections"]:
            if "layout_content" in x:
                if "medias" in x["layout_content"]:
                    for y in x["layout_content"]["medias"]:
                        if y["media"]["media_type"] == 2:
                            ids[y["media"]["code"]] = ""
                            if y["media"].get("video_subtitles_uri"):
                                ids[y["media"]["code"]] = subtitles(requests.get(y["media"]["video_subtitles_uri"]).text)
    except Exception as er:
        print(response.status_code, tag, er)
    
```

```python
with open("data.txt") as file:
    dd = file.readlines()
    for x in dd:
        if len(x) > 1:
            get_tags(x.strip()[1:])
            print(len(ids))
```

```python
with open("vid_data.txt", "a") as file:
    for k, v in ids.items():
        file.write(f"{k}|||{v}\n")
    
```
